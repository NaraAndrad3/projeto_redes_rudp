## Analises e geração dos gráficos

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

pd.set_option("display.max_columns", None)
```

```python
cores = ["#1F77B4", "#FFD700"]
```

```python
app = pd.read_csv('../results/app_metrics_test_512kb.csv')
net = pd.read_csv('../results/network_metrics_512kb.csv')
app.head()
```

```python
net.head()
```

```python
dados = app.merge(net, on = ['run', 'scenario', 'protocol'], how = 'inner')
dados
```

```python
dados['overhead_bytes'] = (dados['network_total_bytes'] - dados['bytes_expected'])
dados['overhead_ratio'] = (dados['network_total_bytes'] / dados['bytes_expected'])
dados['throughput_ratio'] = (dados['throughput_bps'] - dados['network_throughput_bps'])
```

```python
dados.head()
```

- Throughput médio

```python
summary = (
    dados.groupby(["scenario", "protocol"])
      .agg(
          mean_throughput=("throughput_bps", "mean"),
          std_throughput=("throughput_bps", "std")
      )
      .reset_index()
)

summary
```

```python
summary = (
    dados.groupby(["scenario", "protocol"])
      .agg(
          mean_throughput=("throughput_bps", "mean"),
          std_throughput=("throughput_bps", "std")
      )
      .reset_index()
)

fig = px.bar(
    summary,
    x="scenario",
    y="mean_throughput",
    color="protocol",
    barmode="group",
    error_y="std_throughput",
    title="Throughput médio da aplicação",
    labels={
        "scenario": "Cenário",
        "mean_throughput": "Throughput médio (B/s)",
        "protocol": "Protocolo"
    },
    color_discrete_sequence=cores
)

fig.update_yaxes(type="log")

fig.write_image(
    "../results/figures/throughput_app.png",
)
fig.show()


```

- Tempo médio de transferência por cenário.

```python
summary_time = (
    dados.groupby(["scenario", "protocol"])
      .agg(
          mean_time=("elapsed_time_s", "mean"),
          std_time=("elapsed_time_s", "std")
      )
      .reset_index()
)

summary_time
```

```python
fig = px.bar(
    summary_time,
    x="scenario",
    y="mean_time",
    color="protocol",
    barmode="group",
    error_y="std_time",
    title="Tempo médio de transferência por cenário",
    labels={
        "scenario": "Cenário",
        "mean_time": "Tempo médio (s)",
        "protocol": "Protocolo"
    },
    color_discrete_sequence=cores
)

fig.write_image(
    "../results/figures/tempo_medio_transferencia.png",
)

fig.show()
```

- Retransmissões do R-UDP

```python
rudp = dados[dados["protocol"] == "RUDP"]

summary_retrans = (
    rudp.groupby("scenario")
        .agg(
            mean_retransmissions=("retransmissions", "mean"),
            std_retransmissions=("retransmissions", "std")
        )
        .reset_index()
)

summary_retrans
```

```python
fig = px.bar(
    summary_retrans,
    x="scenario",
    y="mean_retransmissions",
    error_y="std_retransmissions",
    title="Retransmissões médias do protocolo R-UDP",
    labels={
        "scenario": "Cenário",
        "mean_retransmissions": "Retransmissões médias"
    },
    color_discrete_sequence=cores
)

fig.write_image(
    "../results/figures/retransmissoes_rudp.png",
)

fig.show()
```

- Overhead de Rede

```python
dados["overhead_bytes"] = (
    dados["network_total_bytes"]
    - dados["bytes_expected"]
)
```

```python
dados[
    [
        "scenario",
        "protocol",
        "bytes_expected",
        "network_total_bytes",
        "overhead_bytes"
    ]
].head()
```

```python
dados[
    [
        "scenario",
        "protocol",
        "bytes_expected",
        "network_total_bytes",
        "overhead_bytes"
    ]
].head()
```

```python
summary_overhead = (
    dados.groupby(["scenario", "protocol"])
      .agg(
          mean_overhead=("overhead_bytes", "mean"),
          std_overhead=("overhead_bytes", "std")
      )
      .reset_index()
)

summary_overhead
```

```python
fig = px.bar(
    summary_overhead,
    x="scenario",
    y="mean_overhead",
    color="protocol",
    barmode="group",
    error_y="std_overhead",
    title="Overhead médio observado na rede",
    labels={
        "scenario": "Cenário",
        "mean_overhead": "Overhead médio (bytes)",
        "protocol": "Protocolo"
    },
    color_discrete_sequence=cores
)

fig.write_image(
    "../results/figures/overhead_rede.png",

)

fig.show()
```

- Quantidade média de pacotes observados na rede

```python
summary_packets = (
    dados.groupby(["scenario", "protocol"])
      .agg(
          mean_packets=("network_packet_count", "mean"),
          std_packets=("network_packet_count", "std")
      )
      .reset_index()
)

summary_packets
```

```python
fig = px.bar(
    summary_packets,
    x="scenario",
    y="mean_packets",
    color="protocol",
    barmode="group",
    error_y="std_packets",
    title="Pacotes médios observados na rede",
    labels={
        "scenario": "Cenário",
        "mean_packets": "Quantidade média de pacotes",
        "protocol": "Protocolo"
    },
    color_discrete_sequence=cores
)

fig.write_image(
    "../results/figures/pacotes_rede.png",

)

fig.show()
```

- Throughput observado na rede (tcpdump)

```python
summary_network_tp = (
    dados.groupby(["scenario", "protocol"])
      .agg(
          mean_network_tp=("network_throughput_bps", "mean"),
          std_network_tp=("network_throughput_bps", "std")
      )
      .reset_index()
)

summary_network_tp
```

```python
fig = px.bar(
    summary_network_tp,
    x="scenario",
    y="mean_network_tp",
    color="protocol",
    barmode="group",
    error_y="std_network_tp",
    title="Throughput médio observado na rede",
    labels={
        "scenario": "Cenário",
        "mean_network_tp": "Throughput médio da rede (B/s)",
        "protocol": "Protocolo"
    },
    color_discrete_sequence=cores
)

fig.update_yaxes(type="log")

fig.write_image(
    "../results/figures/throughput_rede.png",
)

fig.show()
```

- Aplicação × Rede

```python
comparison = (
    dados.groupby(["scenario", "protocol"])
      .agg(
          app_tp=("throughput_bps", "mean"),
          net_tp=("network_throughput_bps", "mean")
      )
      .reset_index()
)

comparison
```

```python
comparison_long = comparison.melt(
    id_vars=["scenario", "protocol"],
    value_vars=["app_tp", "net_tp"],
    var_name="origem",
    value_name="throughput"
)

comparison_long
```

```python
fig = px.bar(
    comparison_long,
    x="scenario",
    y="throughput",
    color="origem",
    facet_col="protocol",
    barmode="group",
    title="Throughput da aplicação versus throughput observado na rede",
    labels={
        "scenario": "Cenário",
        "throughput": "Throughput (B/s)",
        "origem": "Origem da métrica"
    },
    color_discrete_sequence=cores
)

fig.update_yaxes(type="log")

fig.write_image(
    "../results/figures/aplicacao_vs_rede.png",
)

fig.show()
```

```python
final_table = (
    dados.groupby(["scenario", "protocol"])
      .agg(
          throughput_mean=("throughput_bps", "mean"),
          throughput_std=("throughput_bps", "std"),
          time_mean=("elapsed_time_s", "mean"),
          time_std=("elapsed_time_s", "std"),
          retrans_mean=("retransmissions", "mean"),
          retrans_std=("retransmissions", "std")
      )
      .round(2)
)

final_table
```

```python
final_table.to_csv(
    "../results/final_summary.csv"
)
```
